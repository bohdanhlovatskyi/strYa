```python
import pandas as pd
import numpy as np
import seaborn as sns
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

#### -- gettting the data --

```python
def each_sensor_lmplot(path_to_file: str, range_limit: int) -> None:
    df = pd.read_csv(path_to_file)
    df['time_tick'] = range(df.shape[0])
    print(df.head())
    for elm in ['x1', 'y1', 'z1', 'x2', 'y2', 'z2']:
        g = sns.lmplot(x='time_tick', y=elm, data=df, sharey=False)
        if range_limit is not None:
            g.set(ylim=(-range_limit,range_limit))
```

#### -- plotting out the data --

```python
# each_sensor_lmplot('datasets/steady/angles_steady.csv', 90)
each_sensor_lmplot('datasets/walking/angles_walking.csv', 90)
```
